```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,1024,2)
bubble_comp = []
selection_comp = []
insertion_comp = []

for line in open('comp.txt','r'):
    values = [int(v) for v in line.split()]
    bubble_comp.append(values[0])
    selection_comp.append(values[1])
    insertion_comp.append(values[2])

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(16,4))

ax[0].plot(x, bubble_comp)
ax[1].plot(x, selection_comp)
ax[2].plot(x, insertion_comp)

ax[0].set_title('Bubble Sort')
ax[0].grid()

ax[1].set_title('Selection Sort')
ax[1].grid()

ax[2].set_title('Insertion Sort')
ax[2].grid()

plt.show()

plt.figure(figsize=(16,7))
plt.plot(x, bubble_comp, label="Bubble Sort")
plt.plot(x, selection_comp, label="Selection Sort")
plt.plot(x, insertion_comp, label="Insertion Sort")
plt.xlabel("Array Size")
plt.ylabel("Number of comparisons")
plt.title("Number of comparisons in each type of sorting algorithm")
plt.legend()
plt.grid()

plt.show()
```

```python
x = np.arange(0,1024,2)
bubble_exch = []
selection_exch = []
insertion_exch = []

for line in open('exch.txt','r'):
    values = [int(v) for v in line.split()]
    bubble_exch.append(values[0])
    selection_exch.append(values[1])
    insertion_exch.append(values[2])


fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(16,4))

ax[0].plot(x, bubble_exch)
ax[1].plot(x, selection_exch)
ax[2].plot(x, insertion_exch)

ax[0].set_title('Bubble Sort')
ax[0].grid()

ax[1].set_title('Selection Sort')
ax[1].grid()

ax[2].set_title('Insertion Sort')
ax[2].grid()

plt.show()

plt.figure(figsize=(16,7))
plt.plot(x, bubble_exch, label='Bubble Sort')
plt.plot(x, selection_exch, label='Selection Sort')
plt.plot(x, insertion_exch, label='Insertion Sort')
plt.xlabel("Array Size")
plt.ylabel("Number of exchanges")
plt.title('Number of exchanges in each type of sorting algorithm')
plt.legend()
plt.grid()

plt.show()
```

Conditions to use these algorithms:
* _**Bubble Sort**_ is best used to check if an array is sorted or not as it iterates through the array only once. If it takes more than a certain number of comparisons, then the array is not sorted.
* _**Selection Sort**_ is best used for large arrays as it performs the least number of exchanges thereby saving space and time.
* _**Insertion Sort**_ is best used when the size of array is unknown as it gives best results averagely.
