Calculate the linear convolution for the following

## Question 2

Find the output of the system if input x(n) and impulse reaction h(n) are given by:

x(n) = 1    n = -2, 0, 1
     = 2    n = -1
     = 0    otherwise
           
h(n) = \delta(n) − $\delta$(n − 1) + $\delta$(n − 2) − $\delta$(n − 3)


```python
n = np.arange(-2, 5)

x = []
for i in n:
    if i in [-2, 1, 0]:
        x.append(1)
    elif i == -1:
        x.append(2)
    else:
        x.append(0)
        
def impulse(n):
    h = []
    for i in n:
        if i == 0:
            h.append(1)
        else:
            h.append(0)
    return np.array(h)

h = impulse(n) - impulse(n-1) + impulse(n-2) - impulse(n-3)

y = np.trim_zeros(np.convolve(x, h))

plt.figure()
plt.stem(n, y, use_line_collection=True)
plt.grid()
plt.title('Linear Convolution')
plt.xlabel('n')
plt.ylabel('y(n)')
plt.show()
```
