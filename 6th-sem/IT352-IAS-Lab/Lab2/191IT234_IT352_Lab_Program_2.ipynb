```python
# Task 1

import hashlib
import os
import time

start = time.time()
signature_file = open("191IT234_IT352_image_signature_database.txt", "w")

for root, dirs, files in os.walk("Fire images"):
    for file in files:
        with open(os.path.join(root, file), "rb") as img:
            img_bytes = img.read()
            img_hash = hashlib.sha256(img_bytes).hexdigest()
            print(img_hash)
            signature_file.write(img_hash + "\n")

signature_file.close()
print(time.time() - start)
```

```python
# Task 2
# I am assuming that the test case images are stored in the directory "Test-cases"
# The output files are stored in the directory "Test-cases-output"

import os

from IPython.display import Image

i = 1

if not os.path.exists("Test-cases-output"):
    os.makedirs("Test-cases-output")

with open("191IT234_IT352_image_signature_database.txt", "r") as signature_file:
    signatures = signature_file.read().split()

for root, dirs, files in os.walk("Test-cases"):
    for file in sorted(files):
        output_file = open(os.path.join("Test-cases-output", f"191IT234_IT352_Test_Case_{i}.txt"), "w")
        with open(os.path.join(root, file), "rb") as img:
            img_bytes = img.read()
            img_hash = hashlib.sha256(img_bytes).hexdigest()
        display(Image(filename=os.path.join(root, file)))
        print(img_hash)
        print("image exists" if img_hash in signatures else "image does not exist")
        output_file.write(img_hash + "\n")
        output_file.write("image exists" if img_hash in signatures else "image does not exist")
        output_file.close()
        i += 1
```
